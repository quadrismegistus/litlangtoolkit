```python
def add_texts_to_wiki(C,lim=None,shuffle=True,slice_len=10,progress=True):
    texts = C.texts()
    texts = random.sample(texts,lim) if lim and shuffle else texts[:lim]

    iterr = list(slice(texts,slice_length=slice_len))
    if progress:
        from tqdm.notebook import tqdm as tqdmx
        iterr = tqdmx(iterr, desc='Adding texts to wiki in batches', position=0)
    
    for slicex in iterr:
        for i,t in enumerate(slicex):
            if progress:
                qstr=wikidata_query_str(t)
                desc=f'[{i+1:02}/{len(slicex):02}] Querying wikidata: {qstr}'
            res = t.wikidata.metadata(allow_http=True,verbose=False)
            match = '✅' if res.get('qid') else '❌'
            if progress: iterr.set_description(f' {match} {desc}')

        Wiki.save_metadata()
```
