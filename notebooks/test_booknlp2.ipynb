## Character IDs

```python
def get_df_names(self,force=False):
    if force or self._df_names is None:
        char_toks=self.get_character_token_counts()
        o_text=[]
        for char_tok,char_tok_count in char_toks.most_common():
            if type(char_tok)!= str or not char_tok: continue
            char_id=zeropunc(char_tok)
            otextd=dict(char_tok=char_tok,char_id=char_id,char_tok_count=char_tok_count,char_tok_i=len(o_text)+1)
            o_text.append(otextd)
        odf_tok2id=pd.DataFrame(o_text).sort_values(
            ['char_tok_count','char_tok_i'],
            ascending=[False,True]
        ).fillna('').set_index('char_tok')
        odf_tok2id['is_valid']=''
        self._df_names=odf_tok2id
        #save_df(odf_tok2id, self.path_names, verbose=False)
        #self._df_names=merge_read_dfs_anno(self.path_names, on='char_tok').fillna('')
    return self._df_names
```
