```python
def graph(self,text_iter,draw=False):
    g = nx.Graph()
    for t in text_iter:
        for t2 in t.dsources:
            g.add_edge(t.addr, t2.addr)
    
    
    from lltk.model.networks import draw_nx
    return g if not draw else draw_nx(g)

# BaseLLTK.graph = graph
```
